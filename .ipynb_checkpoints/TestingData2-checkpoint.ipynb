# Testing Data Parsing and Visualization

This notebook is to test the different data parsing and visualization options for the experiments

## Load Data and Parse

**Imports and Configuration**

```python
import re
from pathlib import Path

import pandas as pd
import matplotlib.pyplot as plt

# for larger plots
plt.rcParams["figure.figsize"] = (8,4)
```

**Load and pre-filter the raw logs**

```python
log_path = Path("enhanced_rawlog.txt")

with log_path.open() as f:
    raw_lines = [line.rstrip("\n") for line in f]

# Keep only lines from AUDIT_MON (skip monitor headers, etc.)
data_lines = [ln for ln in raw_lines if "AUDIT_MON:" in ln]
len(data_lines), data_lines[:5]
```

**Define parsers for events and markers**

```python
# Base pattern: level, time, module, body
base_pattern = re.compile(
    r'^(?P<level>[WI])\s*'
    r'\((?P<time_ms>\d+)\)\s*'
    r'(?P<module>[A-Z_]+):\s*'
    r'(?P<body>.*)$'
)

# Marker body: MARKER: trial_id=1, scenario_id=S2A, t_change_ms=245000
marker_pattern = re.compile(
    r'^MARKER:\s*trial_id=(?P<trial_id>\d+),\s*'
    r'scenario_id=(?P<scenario_id>[A-Za-z0-9]+),\s*'
    r't_change_ms=(?P<t_change_ms>\d+)\s*$'
)

# AuditValue before/after
audit_pattern = re.compile(
    r'AuditValue\s+(?P<audit_before>-?\d+)->(?P<audit_after>-?\d+)'
)


# PID before/after
pid_pattern = re.compile(
    r'Kp\s+(?P<Kp_before>[\d.]+)->(?P<Kp_after>[\d.]+),\s*'
    r'Ki\s+(?P<Ki_before>[\d.]+)->(?P<Ki_after>[\d.]+),\s*'
    r'Kd\s+(?P<Kd_before>[\d.]+)->(?P<Kd_after>[\d.]+)'
)

# auth flag
auth_pattern = re.compile(r'auth=(?P<auth>[01])')
```

**Define Function**

```python
def parse_line(line: str) -> dict | None:
    m = base_pattern.match(line)
    if not m:
        return None

    d = m.groupdict()
    d["time_ms"] = int(d["time_ms"])

    body = d.pop("body").strip()

    # 1) MARKER lines: encode trial/scenario/t_change_ms
    mm = marker_pattern.match(body)
    if mm:
        d["event_type"] = "MARKER"
        d["trial_id"] = int(mm.group("trial_id"))
        d["scenario_id"] = mm.group("scenario_id")
        d["t_change_ms"] = int(mm.group("t_change_ms"))
        d["auth"] = None
        d["authorized"] = None
        return d

    # 2) Normal events: split "EVENT_TYPE: details"
    if ":" in body:
        evt, details = body.split(":", 1)
        event_type = evt.strip()
        details = details.strip()
    else:
        event_type = body.strip()
        details = ""

    d["event_type"] = event_type

    # AuditValue before/after
    am = audit_pattern.search(details)
    if am:
        d["audit_before"] = int(am.group("audit_before"))
        d["audit_after"] = int(am.group("audit_after"))

    # PID before/after
    pm = pid_pattern.search(details)
    if pm:
        for key in [
            "Kp_before", "Kp_after",
            "Ki_before", "Ki_after",
            "Kd_before", "Kd_after",
        ]:
            d[key] = float(pm.group(key))

    # auth flag
    authm = auth_pattern.search(details)
    if authm:
        d["auth"] = int(authm.group("auth"))
        d["authorized"] = bool(d["auth"])
    else:
        d["auth"] = None
        d["authorized"] = None

    return d
```

## Create Data Tables

**Parse all lines into a DataFrame**

```python
records = []
for ln in data_lines:
    rec = parse_line(ln)
    if rec:
        records.append(rec)

df = pd.DataFrame(records)
df["time_s"] = df["time_ms"] / 1000.0

df.head()
```

**Split into events vs markers**

```python
df_markers = df[df["event_type"] == "MARKER"].copy()
df_events = df[df["event_type"] == "MARKER"].copy()

df_markers.sort_values(["trial_id", "scenario_id"])
```

## Visualizations

**Event Timeline (authorized vs unauthorized)**

```python
# Map authorized flag -> numeric y position
y_map = {
    False: 0,
    True: 1,
    None: 2
}

plt.figure()

# Track whether we actually created any labeled points
added_labels = False

for auth_flag, group in df_events.groupby("authorized"):
    if len(group) == 0:
        continue

    y_val = y_map.get(auth_flag, 2)
    label = (
        "authorized" if auth_flag is True
        else "unauthorized" if auth_flag is False
        else "unknown"
    )

    # Plot with label
    plt.scatter(group["time_s"], [y_val] * len(group), marker="o", label=label)
    added_labels = True

plt.xlabel("Time (s)")
plt.yticks([0, 1, 2], ["unauthorized", "authorized", "unknown"])
plt.title("Watchdog events over time")

# Only draw legend IF we actually created labeled artists
if added_labels:
    plt.legend()

plt.tight_layout()
plt.show()
```

**Event types over time**

```python
event_types = df_events["event_type"].unique()
plt.figure()

for i, et in enumerate(event_types):
    grp = df_events[df_events["event_type"] == et]
    plt.scatter(grp["time_s"], [i] * len(grp), s=60, label=et)

plt.yticks(range(len(event_types)), event_types)
plt.xlabel("Time (s)")
plt.title("Event types over time")
plt.tight_layout()
plt.show()
```
